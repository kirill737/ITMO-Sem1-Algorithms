```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
names = ["1d6", "2d6", "3d6", "1d10", "2d10", "3d10", "10d10", "10d100"]
# names = ["10d10"]
for name in names:
    df = pd.read_csv("stat/" + name + ".csv")
    plt.bar(df["value"], df["prob"])
    plt.title(" " + name)
    plt.show()
```
